### ANALYSIS of the Data



This can be done in major three ways:
1. Product Level Analysis(Category and Sub-Category wise)
2. Region Level Analysis
3. Customer Level Analysis

### Lets start with the Product level Analysis.
Broadly, it is divide into two major domains- Category and Sub-Category.



Lets, first check what are the unique Categories in the dataset, plotting the same with Number of such products by counting the respective values.

```python
df_state['Category'].unique()
```

```python
df_state['Category'].value_counts().plot(kind='bar', figsize=(5,5))
plt.title('Category wise Distribution')
plt.xlabel('Categories')
plt.ylabel('No. of products')
plt.show()
```

With the above plot, we can easily conclude the highest number of products sold from the supermarket was of the Office Suppliers Category.
Let's now check which sub category had the highest sale.

For that we need to know what are the unique Sub-Categories present in the Dataset.

```python
df_state['Sub-Category'].unique()
```

```python
a=df_state['Sub-Category'].value_counts()
```

```python
a.plot.pie(autopct="%1.1f%%",figsize=(10,10))
plt.title('Distribution of Sub-Categotries')
plt.show()
```

This can be clearly seen, the Copiers and Machines have the least number of products sold where as Papers and Binders had the highest Number of products sold.



Now, to have even more clear picture, lets now plot the Sub-Categories inside each Category with the Sales and Profit Gained.

```python
df_c1= df_state.loc[df_state['Category']=='Office Supplies']
df_c2=df_state.loc[df_state['Category']=='Furniture']
df_c3=df_state.loc[df_state['Category']=='Technology']
a=df_c1.groupby('Sub-Category')[['Sales','Profit']].agg(['sum'])
b=df_c2.groupby('Sub-Category')[['Sales','Profit']].agg(['sum'])
c=df_c3.groupby('Sub-Category')[['Sales','Profit']].agg(['sum'])
```

```python
fig=plt.figure()
ax0=fig.add_subplot(1,3,1)
ax1=fig.add_subplot(1,3,2)
ax2=fig.add_subplot(1,3,3)

a.plot(kind='bar', ax=ax0,figsize=(8,8))
ax0.set_title('Office Supplies')
ax0.set_xlabel('sub-Category')
ax0.set_ylabel('Sales and Profit')
b.plot(kind='bar', ax=ax1,figsize=(8,8))
ax1.set_title('Furniture')
ax1.set_xlabel('sub-Category')
ax1.set_ylabel('Sales and Profit')
c.plot(kind='bar', ax=ax2,figsize=(8,8))
ax2.set_title('Technology')
ax2.set_xlabel('sub-Category')
ax2.set_ylabel('Sales and Profit')
fig.tight_layout()                  
plt.show()
```

As per the above plot, it is well understood that the sale of Fasteners was least among all the three Sub-Categories, and that of Chairs and Phones were highest and the profit from Copiers is highest among all, where as the loss by Tables is highest.

### Lets now work on the State and Region Level Analysis.


First of all, check how the Sale Trend goes with respect to the State.

```python
df_state.groupby('State')[['State','Sales']].agg(['sum']).plot(kind='bar',figsize=(20,5))
plt.ylabel('Sale')
plt.title('Sale in various States')
plt.show()
```

The highest sale was in California and New york, the Texas and Washington had the moderate Sale and the other state had comparably low Sale.

```python
df_state.groupby('State')[['State','Profit']].agg(['sum']).plot(kind='bar',figsize=(20,8),color='g');
plt.ylabel('Profit or Gain')
plt.title('Profit made by the Superstore in various States')
plt.show()
```

The Sale and Profit are directly propotional to each other, the same can be seem here, the States with the highest sale i.e., California and New york shows the cities where the Supermarket earned the highest Profit. The Texas being an exception case shows a higher loss and Washington also had a good profit when compared to the other states.

Lets now see the Sale and Profit Trend with respect to the different regions in the Country.

```python
df_state.groupby('Region')[['Region','Sales']].agg(['sum']).plot(kind='bar',figsize=(20,5))
plt.ylabel('Sale')
plt.title('Sale in various Regions')
plt.show()
```

```python
df_state.groupby('Region')[['Region','Profit']].agg(['sum']).plot(kind='bar',figsize=(20,5),color='y')
plt.ylabel('Profit or Gain')
plt.title('Profit made by the Superstore in various Regions')
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x="Sub-Category", hue="Region", data=df_state)
plt.title('Distrubution of sub-categories Region wise')
plt.show()
```

With above three plots, the inference be taken out the West Region of the Country is the region where supermarket had the highest as well as profit, the central region had more sale then Souhtern region but had lesser Profit from it.



### Lets now work on the Segment or Consumer Level Analysis.


We have to check how the Sale and Profit trend goes with Segment.

```python
fig=plt.figure()
ax0=fig.add_subplot(1,2,1)
ax1=fig.add_subplot(1,2,2)
df_state.groupby('Segment')[['Segment','Sales']].agg(['sum']).plot(kind='bar',ax=ax0,figsize=(10,5),color = 'b');
ax0.set_title('Segment V/S Sales')
ax0.set_xlabel('Segment')
ax0.set_ylabel('Sales')
df_state.groupby('Segment')[['Segment','Profit']].agg(['sum']).plot(kind='bar',ax=ax1,figsize=(10,5), color='r');
ax1.set_title('Segment V/S ')
ax1.set_xlabel('Segment')
ax1.set_ylabel('Profit')
fig.tight_layout()
plt.show()
```

The Consumer domain of the Segment had the highest Sale as well as the profit and home office had the least.
